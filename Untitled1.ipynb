```python
import numpy as np
import re
import scipy.spatial
```

```python
doc = open('sentenves.txt', 'r')
```

```python
doc_list = doc.readlines()
```

```python
doc_lower = []
```

```python
for line in doc_list: doc_lower.append(line.lower())
```

```python
doc_test= []
```

```python
for line in doc_lower:
    l = re.split('[^a-z]', line)
    l_d = []
    for w in l:
        if len(w) != 0:
            l_d.append(w)
    doc_test.append(l_d)
```

```python
doc_test
```

```python
d = dict()
c = 0
for line in doc_test:
    for w in line:
        d[w] = 0
print len(d)
for key in d:
    d[key] = c
    c += 1
d
```

```python
vectors = np.zeros((22, 254))
i = 0
for line in doc_test:
    j = 0
    for w in line:
        vectors[i, j] = d[w]
        j += 1
    i += 1
print vectors
```

```python
for i in range(1, 22):
    print i, scipy.spatial.distance.cosine(vectors[0],vectors[i])
```
